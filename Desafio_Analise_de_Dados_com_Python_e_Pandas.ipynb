```python
# Filtrando dados da coluna selecionada
tabela['ITEM'].unique
```

```python
# Calculando o valor total da coluna
round(tabela["TOTAL"].sum(), 2)
```

```python
# Exibindo dados por coluna
Dados_por_produto = tabela[["ITEM", "QUANTIDADE", "VALOR UNITÁRIO", "TOTAL"]].groupby(['ITEM', 'VALOR UNITÁRIO']).sum()
print(Dados_por_produto)
```

```python
# Exibindo o valor total do ítem
tabela["TOTAL"] = tabela["LUCRO"].mul(tabela["QUANTIDADE"])
tabela.head(10)
```

```python
# Exibe o lucro das vendas
tabela["LUCRO"] = tabela['VALOR UNITÁRIO'].mul(tabela['ACRÉSSIMO %']) + tabela["QUANTIDADE"]
tabela.head(10)
```
